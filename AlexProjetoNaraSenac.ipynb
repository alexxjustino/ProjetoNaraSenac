```python
array_df_reservas = np.array(df_reservas_quarto['valor_diaria'])
q1_df_reservas = np.percentile(array_df_reservas, 25)
q2_df_reservas = np.percentile(array_df_reservas, 50)
q3_df_reservas = np.percentile(array_df_reservas, 75)
media = np.mean(array_df_reservas)
mediana = np.median(array_df_reservas)
distancia = (media - mediana) / mediana

print(f'1º quartil: {q1_df_reservas}')
print(f'2º quartil (mediana): {q2_df_reservas}')
print(f'3º quartil: {q3_df_reservas}')
print(f'média: {media}')
print(f'mediana: {mediana}')
print(f'distância: {round(distancia * 100, 2)}%')
print('\nConclusão')  
print('Podemos usar a média como medida de tendência central uma vez que a distância entre média e mediana é de 5.49%.')
print(f'valor médio das estadias R$ {media:.2f}')
#criando limites superior e inferior
iqr = q3_df_reservas - q1_df_reservas
limite_superior = q3_df_reservas + (iqr * 1.5)
limite_inferior = q3_df_reservas - (iqr * 1.5)
print(f'limite superior: {limite_superior}')
print(f'limite inferior: {limite_inferior:.2f}')
```
